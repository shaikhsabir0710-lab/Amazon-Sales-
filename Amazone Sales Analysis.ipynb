### Data Cleaning

```python
# we checked how many null values present in our dataset per coloumns.
df.isnull().sum()
```

```python
# we can see the new column and pendings cloumn has null column we delete both columns.
df=df.drop(columns=({'New','PendingS'}))
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
# we change 'ship-postal-code' floate datatype into int64 datatype.
df['ship-postal-code']=df['ship-postal-code'].astype('int64')
```

```python
# we can change the date column datatype object into datatime datatype.
df['Date']=df['Date'].astype('datetime64[ns]')
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.dtypes
```

```python
#  Random profit margin assign kara (10% - 40%)
unique_cats = df['Category'].unique()
margins = np.random.uniform(0.10, 0.40, size=len(unique_cats))  # 10% to 40%
margin_dict = dict(zip(unique_cats, margins))

# Step 4: Margin map karna
df['Profit_Margin'] = df['Category'].map(margin_dict)

# Step 5: Profit aur Cost nikalna
df['Profit'] = df['Revenue'] * df['Profit_Margin']
df['Cost'] = df['Revenue'] - df['Profit']
```

```python
# extract month name,year,weekday from date column.
df['Month'] = df['Date'].dt.month_name()
df['Year'] = df['Date'].dt.year
df['Weekday'] = df['Date'].dt.day_name()
```

### EXPLOARY DATA ANALYSIS

```python
ax=sns.countplot(x='Size',data=df)
plt.title('Size Wise Count')
plt.xlabel('SIZE')
plt.ylabel('COUNT')
plt.style.use('_classic_test_patch')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

### Most highest size is 'M' size.

```python
df['Category']=df['Category'].astype(str)
category_profit=df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
bars=plt.bar(category_profit.index,category_profit.values,color='olive')
plt.show()
```

### Most Profitable Category is 'Shirt'

```python
st=df['Status'].value_counts()
plt.figure(figsize=(8,6))
plt.barh(st.index,st.values,color='lightpink')
plt.title('STATUS')
plt.show()
```

```python
state_wise_profit=df.groupby('ship-state')['Profit'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(10,5))
plt.pie(state_wise_profit,labels=state_wise_profit.index,autopct='%1.1f%%',startangle=90)
plt.xticks(rotation=45)
plt.title('Top5 Higest Profitable State')
plt.style.use('_classic_test_patch')
plt.show()
```

### Most Profitable State is 'Maharashtra' =  29.9%.
